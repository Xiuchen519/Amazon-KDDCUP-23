```python
product_train = pd.read_csv('/root/autodl-tmp/xiaolong/WorkSpace/Amazon-KDDCUP-23/raw_data/products_train.csv')
```

```python
product_train.set_index('id', inplace=True)
product_train
```

```python
def compare_title(sessions_df, product_df, prediction_df, index, is_test=False):
    print('********last item*********')
    prev_items = eval(sessions_df.iloc[index]['prev_items'].replace(' ', ','))
    last_items = prev_items[-5:]
    for last_item in last_items:
        last_item_title = product_df.loc[last_item][['title', 'locale']]
        print(last_item_title)
    if not is_test:
        print('*******ground truth*******')
        next_item = sessions_df.iloc[index]['next_item']
        next_item_title = product_df.loc[next_item][['title', 'locale']]
        print(next_item_title)
    print('*******prediction*********')
    for i in range(10):
        print(f'No. {i} : ')
        candidate = prediction_df.iloc[index]['next_item_prediction'][i]
        predict_title = product_df.loc[candidate][['title', 'locale']]
        print(predict_title)
```

```python
compare_title(task3_test_sessions, product_train, task3_test_prediction_df, 55010, is_test=True)
```
